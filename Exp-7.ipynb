```python
from bs4 import BeautifulSoup
import requests 

# Function to extract Product Title
def get_title(soup):
    try:
        title = soup.find("span", attrs={"id": "productTitle"})
        if title is not None:
            title_string = title.get_text(strip=True)
            print("Title found:", title_string)
        else:
            title_string = ""
    except AttributeError:
        title_string = ""
    return title_string

# Function to extract Product Price
def get_price(soup):
    try:
        price = soup.find("span", attrs={"class": "a-price"})
        if price is not None:
            price_string = price.get_text(strip=True)
            print("Price found:", price_string)
        else:
            price_string = ""
    except AttributeError:
        price_string = ""
    return price_string

# Function to extract Product Rating
def get_rating(soup):
    try:
        rating = soup.find("span", attrs={"class": "a-icon-alt"})
        if rating is not None:
            rating_string = rating.get_text(strip=True)
            print("Rating found:", rating_string)
        else:
            rating_string = ""
    except AttributeError:
        rating_string = ""
    return rating_string

# Function to extract Number of User Reviews
def get_review_count(soup):
    try:
        review_count = soup.find("span", attrs={"id": "acrCustomerReviewText"})
        if review_count is not None:
            review_count_string = review_count.get_text(strip=True)
            print("Review count found:", review_count_string)
        else:
            review_count_string = ""
    except AttributeError:
        review_count_string = ""
    return review_count_string

# Function to extract Availability Status
def get_availability(soup):
    try:
        availability = soup.find("div", attrs={"id": "availability"})
        if availability is not None:
            availability_string = availability.get_text(strip=True)
            print("Availability found:", availability_string)
        else:
            availability_string = ""
    except AttributeError:
        availability_string = ""
    return availability_string

if __name__ == "__main__":
    # Headers for request
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36",
        "Accept-Language": "en-US, en;q=0.5"
    }

    # The webpage URL
    url = "https://www.amazon.in/Supports-Rejection-Precise-Writing-Drawing/dp/B09SPRK9QF/ref=sr_1_2_sspa?crid=2HJALKK655PWB&dib=eyJ2IjoiMSJ9.1y1iI2kxr5x0M-uvUk4xnSSBQA5aCiMAIOxmpbbkhCX__xpN9lE3nxekcve4NhU_dxfP9rZ9IXk6YN1GQxDwERSQBdBJu3knisZ40Dt-gTAEVln2SugHgGOrczfIA5R63j4RylITPKukmYtHAS_qRHFmO7uEljS9H3lD4W2eH2_ur9jFeGZ-v0G0AHeZ-84pCkABtT8Tj7lrPcYMoUPCreWD48gBrPgQRxsQcQkzJpM.GDVSPKntge6GuVlZMcf28kAOsdNDjC5U0aiAxm2RABk&dib_tag=se&keywords=Stylus&qid=1715604635&sprefix=stylus%2Caps%2C217&sr=8-2-spons&sp_csd=d2lkZ2V0TmFtZT1zcF9hdGY&psc=1"
   
    # HTTP Request
    webpage = requests.get(url, headers=headers)

    # Soup Object containing all data
    soup = BeautifulSoup(webpage.content, "html.parser")

    # Print out the HTML content for debugging
    print(soup.prettify())

    # Function calls to display all necessary product information
    print("Product Title =", get_title(soup))
    print("Product Price =", get_price(soup))
    print("Product Rating =", get_rating(soup))
    print("Number of Product Reviews =", get_review_count(soup))
    print("Availability =", get_availability(soup))
```
